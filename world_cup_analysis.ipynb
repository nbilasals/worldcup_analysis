```python
import pandas as pd, numpy as np, os, seaborn as sns, plotly.express as px, matplotlib.pyplot as plt,plotly.graph_objects as go
from plotly.offline import init_notebook_mode
from wordcloud import WordCloud
```

```python
wc_summ = pd.read_csv("/content/drive/MyDrive/dataset/project/world cup analysis/FIFA - World Cup Summary.csv")
wc_summ
```

```python
wc_summ.loc[wc_summ["CHAMPION"] == 'West Germany', "CHAMPION"] = 'Germany'
wc_summ.loc[wc_summ["HOST"] == 'West Germany', "HOST"] = 'Germany'
wc_summ.loc[wc_summ["RUNNER UP"] == 'West Germany', "RUNNER UP"] = 'Germany'
wc_summ.loc[wc_summ["THIRD PLACE"] == 'West Germany', "THIRD PLACE"] = 'Germany'
```

```python
datafiles =[file for file in os.listdir('/content/drive/MyDrive/dataset/project/world cup analysis/fifa-football-world-cup-dataset')]
datafiles
```

```python
fifa_WCs = pd.DataFrame()
for file in datafiles:
    if file != 'FIFA - World Cup Summary.csv':
        df = pd.read_csv('/content/drive/MyDrive/dataset/project/world cup analysis/fifa-football-world-cup-dataset/' + file)
        fifa_WCs = pd.concat([fifa_WCs, df])
fifa_WCs
```

```python
fifa_WCs.loc[fifa_WCs["Team"] == 'West Germany', "Team"] = 'Germany'
```

```python
fifa_WCs_teams = fifa_WCs.groupby(['Team']).sum()
fifa_WCs_teams
```

```python
fifa_WCs_teams['Win Rate'] = (fifa_WCs_teams['Win']/fifa_WCs_teams['Games Played'])*100
fifa_WCs_teams
```

```python
fifa_WCs_teams = fifa_WCs_teams.sort_values(['Games Played','Win'], ascending = False)
fifa_WCs_teams
```

```python
# Select the top 10 teams with the most appearances
top_10_teams = fifa_WCs_teams.nlargest(10, 'Games Played')

fig = px.bar(top_10_teams, x=top_10_teams.index, y='Games Played',
             labels={'index': 'Teams', 'Games Played': 'Games Played'},
             title='Top 10 Teams with the Most FIFA World Cup Appearances')
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
fifa_WCs_teams = fifa_WCs_teams.sort_values('Win', ascending=False)

# Select the top 10 teams with the most wins
top_10_teams = fifa_WCs_teams.head(10)

fig = px.bar(top_10_teams, x=top_10_teams.index, y='Win',
             labels={'index': 'Teams', 'Win': 'Wins'},
             title='Top 10 FIFA World Cup Teams with the Most Wins')

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
# Sort the DataFrame by 'Win Rate' in descending order
fifa_WCs_teams = fifa_WCs_teams.sort_values('Win Rate', ascending=False)

top_10_teams = fifa_WCs_teams.head(10)

fig = px.bar(top_10_teams, x=top_10_teams.index, y='Win Rate',
             labels={'index': 'Teams', 'Win Rate': 'Win Rate'},
             title='Top 10 FIFA World Cup Teams with the Highest Win Rates')

fig.update_layout(xaxis_tickangle=-45)

fig.show()
```

```python
import plotly.express as px

# Count the number of runner-up appearances for each team
table_RUNNER_UP_wc_summ = wc_summ['RUNNER UP'].value_counts()

# Sort the DataFrame by the count of runner-up appearances in descending order
table_RUNNER_UP_wc_summ = table_RUNNER_UP_wc_summ.reset_index()
table_RUNNER_UP_wc_summ.columns = ['Team', 'Count']
table_RUNNER_UP_wc_summ = table_RUNNER_UP_wc_summ.sort_values('Count', ascending=False)

fig = px.bar(table_RUNNER_UP_wc_summ, x='Team', y='Count',
             labels={'Team': 'Teams', 'Count': 'Number of Runner-Up Appearances'},
             title='Number of World Cup Runner-Up Appearances per Team')
fig.update_xaxes(categoryorder='total descending', tickangle=45)
fig.show()
```

```python
table_THIRD_PLACE_wc_summ = wc_summ['THIRD PLACE'].value_counts()

table_THIRD_PLACE_wc_summ = table_THIRD_PLACE_wc_summ.reset_index()
table_THIRD_PLACE_wc_summ.columns = ['Team', 'Count']
table_THIRD_PLACE_wc_summ = table_THIRD_PLACE_wc_summ.sort_values('Count', ascending=False)

fig = px.bar(table_THIRD_PLACE_wc_summ, x='Team', y='Count',
             labels={'Team': 'Teams', 'Count': 'Number of Third-Place Finishes'},
             title='Number of World Cup Third-Place Finishes per Team')

fig.update_xaxes(categoryorder='total descending', tickangle=45)
fig.show()
```

```python
appearences = fifa_WCs['Team'].value_counts().rename('appearences')

fifa_WCs_appearences = fifa_WCs.merge(appearences.to_frame(),
                                left_on='Team',
                                right_index=True)
```

```python
fifa_WCs_appearences
```

```python
fifa_WCs_appearences_top_10 = fifa_WCs_appearences[fifa_WCs_appearences.appearences > 13]
fifa_WCs_appearences_top_10
```

```python
fifa_WCs_appearences_top_10['Team'].nunique()
```

```python
table_Team_fifa_WCs_appearences_top_10 = fifa_WCs_appearences_top_10['Team'].value_counts()

# Sort the DataFrame by the count of appearances in descending order
table_Team_fifa_WCs_appearences_top_10 = table_Team_fifa_WCs_appearences_top_10.reset_index()
table_Team_fifa_WCs_appearences_top_10.columns = ['Team', 'Count']
table_Team_fifa_WCs_appearences_top_10 = table_Team_fifa_WCs_appearences_top_10.sort_values('Count', ascending=False)

fig = px.bar(table_Team_fifa_WCs_appearences_top_10, y='Team', x='Count',
             labels={'Team': 'Teams', 'Count': 'Number of World Cup Appearances'},
             title='Top 10 Teams with the Most World Cup Appearances')
fig.update_yaxes(categoryorder='total ascending', tickangle=0)
fig.show()
```
